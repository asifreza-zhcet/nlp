## Detect languages

### lang_detect libary

```python
from langdetect import detect_langs
texts = ['Dobrý den, rád tě vidím', 'Konnichiwa, genki desu ka?',
'Bonjour, comment allez-vous ?', 'Habari gani?','مرحبا كيف حالك ', 'Jambo, habari yako?',
'Dia dhuit, cianḟúara', 'Здравствуйте, как поживаете?', 'Ciao, come stai?', 'Hallo, wie geht es dir?']
for text in texts:
    print(f'{text} : [{detect_langs(text)}]')
```

## googletrans

```python
import googletrans
print(googletrans.LANGUAGES)
```

## Translate

### Translate to odiya

```python
from translate import Translator
tr = Translator(to_lang = 'or')
for text in texts:
    print(f'{text}---to----odiya---->{tr.translate(text)}')
```

### Translate to Telugu

```python
from translate import Translator
tr = Translator(to_lang = 'te')
for text in texts:
    print(f'{text}---to----telugu---->{tr.translate(text)}')
```
